```python
from selenium import webdriver
import time
from datetime import datetime
from selenium.webdriver import Edge
import pandas as pd
```

```python
names_all = []
brands_all = []
prices_all = []
urls_all = []
images_all = []

driver = Edge(executable_path = 'c:/users/em17/desktop/edgedriver_win64/msedgedriver.exe')
driver.get('https://www.notino.co.uk/toothpaste/?npc=20')
time.sleep(3)
page_products = driver.find_elements_by_class_name('sc-jdHILj.OFtqG')
for product in page_products:
    name = product.find_element_by_class_name('sc-dmyCSP.sc-ftxyOh.eDlssm.icLilU').text
    names_all.append(name)
    brand = product.find_element_by_class_name('sc-guDLey.sc-jPpdYo.kbBsIA.dloLns').text
    brands_all.append(brand)
    try:
        price = product.find_element_by_class_name('sc-hhyKGa.sc-gYrqIg.iwwcvf.dOVzXY').text
        prices_all.append(price)
    except:
        prices_all.append(None)
    url = product.get_attribute('href')
    urls_all.append(url)
    image = product.find_element_by_tag_name('img').get_attribute('src')
    images_all.append(image)
driver.quit()
```

```python
import pandas as pd
```

```python
notino_products = pd.DataFrame({'name': names_all, 'brand': brands_all, 'price': prices_all, 'url': urls_all, 'image': images_all})
```

```python
notino_products.to_csv('notino_raw.csv')
```
